Information Sheet for USER
----------------------------

1. Simulation Details :

    omega_m = 0.308
    omega_l = 0.692
    h = 0.678
    omega_b = 0.04
    sigma_8 = 0.829
    ns = 0.961
    YHe = 0.24

    Box Length = 160 cMpc/h

    Grid Cell Length = 2 cMpc/h

    No of grids along each direction (ngrid) = 80 

    So, there are 80^3 grid cells in total 

2. Redshift at which files are shared :

    z = 9.0 
    Mass Averaged Ionized Fraction =  0.162
    Number of galaxies  = 189692

    z = 8.0 
    Mass Averaged Ionized Fraction =  0.303
    Number of galaxies  = 420462

    z = 7.0 
    Mass Averaged Ionized Fraction =  0.543
    Number of galaxies  = 830634

    z = 6.0 
    Mass Averaged Ionized Fraction =  0.927
    Number of galaxies  = 1459007

    Mass Averaged Ionized fraction = Mean of " QHII_cell * ( 1 + densitycontrast_cell) "  over all cells 
    where QHII_cell is the ionized fraction and densitycontrast_cell is the density contrast of the cell 

 
3. Description of individual files shared for each redshift bin :

    A. Ionized fraction field 

    Contains the ionization fraction in each cell of the simulation box (unitless) : qion_arr 


    B. Density Contrast field  

    Contains the density contrast in each cell of the simulation box (unitless) : densitycontrast_arr 


    C. Galaxy Catalogue

    (i) galaxyCatalog_UVmagnitudes_z***.**.npz 

        Contains the UV magnitude of each galaxy (units= AB mag) : MUV_arr

    (ii) galaxyCatalog_Xposition_z***.**.npz 

        Contains the X-coordinate position of each galaxy (units= cMpc/h) : xpos_arr 

    (iii) galaxyCatalog_Yposition_z***.**.npz 

        Contains the Y-coordinate position of each galaxy (units= cMpc/h) : ypos_arr 

    (iv) galaxyCatalog_Zposition_z***.**.npz 

        Contains the Z-coordinate position of each galaxy (units= cMpc/h) : zpos_arr 

    So for the Mth galaxy (i.e., array_index = M-1), 
    the set of its properties are MUV_arr[M-1] ; xpos_arr[M-1] ; ypos_arr[M-1] ; zpos_arr[M-1]

4. Other Information :

    A seperate folder named "extra_plots" contains some visualization plots, such as : 

         a. The luminosity function of the galaxies in the mock catalog at different redshifts  
         b. Dimmensionless cross-power spectrum of the mass-averaged neutral hydrogen field and galaxy field at different redshifts. [Definitions used for the calculation are given in a seperate image file] 
    

### Importing

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Load data and plot galaxies with -15 <= MUV <= -14.99

```python
# Load the data
data_dir = '/Users/kwanitgangopadhyay/Desktop/Reionization kNN/GADGET-2_Runs/files_z006.00/'
xpos_data = np.load(data_dir + 'galaxyCatalog_Xposition_z006.00.npz')
ypos_data = np.load(data_dir + 'galaxyCatalog_Yposition_z006.00.npz')
zpos_data = np.load(data_dir + 'galaxyCatalog_Zposition_z006.00.npz')
muv_data = np.load(data_dir + 'galaxyCatalog_UVmagnitudes_z006.00.npz')

# Extract arrays
xpos_arr = xpos_data['xpos_arr']
ypos_arr = ypos_data['ypos_arr']
zpos_arr = zpos_data['zpos_arr']
muv_arr = muv_data['MUV_arr']

# Combine coordinates into a single array
positions = np.vstack((xpos_arr, ypos_arr, zpos_arr)).T  # Shape (N, 3)

# Apply magnitude cut (-15 <= MUV <= -14.99)
magnitude_min = -15
magnitude_max = -14.99
valid_indices = (muv_arr >= magnitude_min) & (muv_arr <= magnitude_max)
filtered_positions = positions[valid_indices]

# Print basic information
print(f"Total galaxies: {len(positions)}")
print(f"Galaxies after magnitude cut ({magnitude_min} <= MUV <= {magnitude_max}): {len(filtered_positions)}")

# 3D Scatter Plot of Galaxy Positions
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(filtered_positions[:, 0], filtered_positions[:, 1], filtered_positions[:, 2], 
           c='blue', s=1, alpha=0.5)

# Set axis labels
ax.set_xlabel('X (cMpc/h)')
ax.set_ylabel('Y (cMpc/h)')
ax.set_zlabel('Z (cMpc/h)')
ax.set_title(f'3D Galaxy Positions at z = 6.0\n({magnitude_min} <= MUV <= {magnitude_max})')

plt.show()
```

### Load data and plot ionized fraction field

```python
# Constants
ngrid = 80  # Number of grid points along each axis (from the info sheet)
grid_cell_size = 2  # Grid cell size in cMpc/h
box_length = ngrid * grid_cell_size  # Total box length in cMpc/h

# Load the ionized fraction data
ionized_data = np.load(data_dir + 'ionizedfraction_ngrid0080_z006.00.npz')  # Replace with the correct filename if necessary
qion_arr = ionized_data['qion_arr']  # Ionized fraction array

# Reshape the data to a 3D grid
qion_grid = qion_arr.reshape((ngrid, ngrid, ngrid))  # Reshape into a cubic 3D grid

# Define slice index
slice_index = ngrid // 2  # Take the middle slice for visualization

# Extract a 2D slice along the Z-axis
qion_slice = qion_grid[:, :, slice_index]

# 2D Plot of the Ionized Fraction with Physical Lengths
plt.figure(figsize=(8, 6))
plt.imshow(
    qion_slice,
    origin='lower',
    cmap='plasma',
    extent=[0, box_length, 0, box_length]  # Physical lengths in cMpc/h
)
plt.colorbar(label='Ionized Fraction (QHII)')
plt.title(f'Ionized Fraction Field Slice (Z = {slice_index}) at z = 6.0')
plt.xlabel('X (cMpc/h)')
plt.ylabel('Y (cMpc/h)')

plt.show()

# Loop through slices to generate multiple plots
for z in range(0, ngrid, 10):  # Plot every 10th slice
    plt.figure(figsize=(8, 6))
    plt.imshow(
        qion_grid[:, :, z],
        origin='lower',
        cmap='plasma',
        extent=[0, box_length, 0, box_length]  # Physical lengths in cMpc/h
    )
    plt.colorbar(label='Ionized Fraction (QHII)')
    plt.title(f'Ionized Fraction Field Slice (Z = {z}) at z = 6.0')
    plt.xlabel('X (cMpc/h)')
    plt.ylabel('Y (cMpc/h)')
    plt.show()
```

### Packing all the galaxy information into one single object

```python
import numpy as np

redshifts = [6, 7, 8, 9]

for z in redshifts:
    data_dir = f'/Users/kwanitgangopadhyay/Desktop/Reionization kNN/GADGET-2_Runs/files_z00{z}.00/'

    # Load galaxy data
    xpos_data = np.load(data_dir + f'galaxyCatalog_Xposition_z00{z}.00.npz')
    ypos_data = np.load(data_dir + f'galaxyCatalog_Yposition_z00{z}.00.npz')
    zpos_data = np.load(data_dir + f'galaxyCatalog_Zposition_z00{z}.00.npz')
    muv_data = np.load(data_dir + f'galaxyCatalog_UVmagnitudes_z00{z}.00.npz')

    xpos_arr = xpos_data['xpos_arr']
    ypos_arr = ypos_data['ypos_arr']
    zpos_arr = zpos_data['zpos_arr']
    muv_arr = muv_data['MUV_arr']  # Galaxy Magnitudes
    
    # Combine positions into a single (N,3) array
    positions = np.vstack((xpos_arr, ypos_arr, zpos_arr)).T  

    # Save galaxies (positions and MUV)
    np.savez_compressed(
        data_dir + f'galaxies_z{z}.npz',
        positions=positions,  # Shape (N,3)
        muv=muv_arr  # Shape (N,)
    )
    print(f"Saved galaxy data for z={z} as {data_dir}galaxies_z{z}.npz")

    # Load ionized fraction and density contrast data
    ionized_data = np.load(data_dir + f'ionizedfraction_ngrid0080_z00{z}.00.npz') 
    qion_arr = ionized_data['qion_arr']  # Ionized fraction array
    
    density_contrast = np.load(data_dir + f'densitycontrast_ngrid0080_z00{z}.00.npz')  
    del_m = density_contrast['densitycontrast_arr']  # Matter overdensity field
    
    # Compute Mass Averaged Ionized Fraction Overdensity Field
    one = np.ones_like(del_m)
    rho_HI = qion_arr * (one + del_m)
    del_HI = rho_HI / np.mean(rho_HI) - one  # Ionized fraction overdensity field

    # Save fields (qion, del_m, del_HI)
    np.savez_compressed(
        data_dir + f'fields_z{z}.npz',
        qion=qion_arr,  # Shape (128,128,128)
        del_m=del_m,  # Shape (128,128,128)
        del_HI=del_HI  # Shape (128,128,128)
    )
    print(f"Saved field data for z={z} as {data_dir}fields_z{z}.npz")
```
